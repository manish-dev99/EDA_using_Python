```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Cardio.csv')
df
```

```python
df.info()
```

```python
sns.set_style('darkgrid')
sns.set_palette('viridis')
sns.set_context('notebook')
sns.lineplot(x='Age',y='Usage',data=df,hue='Gender',marker='o')
plt.show()
```

```python
sns.set_style('darkgrid')
# sns.set_palette({'Male': 'rocket', 'Female': 'green'}) # This caused the error
sns.set_context('paper')
sns.lineplot(x='Age',y='Usage',data=df,hue='Gender',marker='o', palette={'Male': 'red', 'Female': 'green'})
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_palette('RdBu')
sns.set_context('talk')
sns.lineplot(x='Age',y='Usage',data=df,hue='Gender',marker='o')
plt.show()
```

```python
sns.set_style('ticks')
sns.set_palette('vlag')
sns.set_context('poster')
sns.lineplot(x='Age',y='Usage',data=df,hue='Gender',marker='o')
plt.show()
```

```python
plt.figure(figsize=(150,200),dpi=300)
df_filtered = df[df['Gender'] != 'Female'].copy()
g = sns.catplot(x='Age',y='Usage',data=df_filtered,kind='bar',errorbar=None)
g.set_xticklabels(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
g=sns.barplot(x='Age',y='Usage',data=df,hue='MaritalStatus',errorbar=None, color='blue')
g.set_xticklabels(g.get_xticklabels(), rotation=45, ha="right")
```

```python
sns.pairplot(df)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(data=numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()
```

```python
e=sns.stripplot(x='Age',y='Usage',data=df,hue='Gender',jitter=True,color='red')
e.set_xticklabels(e.get_xticklabels(), rotation=45, ha="right")
```

```python
sns.stripplot(x='Product',y='Age',data=df,hue='Gender',jitter=False)
```

```python
product_counts = df['Product'].value_counts()

plt.pie(x=product_counts.values, labels=product_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Products')
plt.show()
```
