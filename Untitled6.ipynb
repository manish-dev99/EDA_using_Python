```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font1 = {'family':'serif','color':'#7FFF00','size':20}
font2 = {'family':'serif','color':'blue','size':15}

plt.plot(x1, y1, x2, y2)
plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data" ,fontdict = font2)


plt.grid()
plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
term_counts = df['Term'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(term_counts, labels=term_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Students by Term')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Student ID'], df['Attendance (%)'], color = 'hotpink' , alpha = 0.8,edgecolors ='green' )
plt.xlabel('Student ID')
plt.ylabel('Attendance (%)')
plt.title('Student Attendance Percentage')
plt.xticks([])  # Hide x-axis labels for clarity with many students
plt.show()
```

```python
term_counts = df['Term'].value_counts()

plt.figure(figsize=(10, 6))
term_counts.plot(kind='bar' , color = 'Red', width =0.2, alpha = 0.2)
plt.xlabel('Term')
plt.ylabel('Number of Students')
plt.title('Number of Students per Term')
plt.xticks(rotation=0)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.4 )
plt.show()
```

```python
term_counts = df['Term'].value_counts()

plt.figure(figsize=(10, 6))
term_counts.plot(kind='barh' , color = 'Green', width =0.4, )
plt.xlabel('Term')
plt.ylabel('Number of Students')
plt.title('Number of Students per Term')
plt.xticks(rotation=0)
plt.grid()
plt.show()
```
