```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('assignment.csv')
```

```python
df2= df.head(25)
```

```python
df2['Premium'] = df2['Premium'].str.replace(',','').astype(int)
plt.figure(figsize=(10, 6))
colors = np.random.rand(len(df))
plt.scatter(df2['Electric add on rate'], df2['Premium'] , c= colors)
plt.title('Scatter plot of Premium vs Electric add on rate')
plt.xlabel('Electric add on rate')
plt.ylabel('Premium')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
colors = np.random.rand(len(df2))
sizes = 10 * np.random.randint(100, size=(100))
plt.scatter(df2['Electric add on rate'], df2['Premium'] , c= colors,alpha= 0.5, cmap= 'nipy_spectral')
plt.title('Scatter plot of Premium vs Electric add on rate')
plt.xlabel('Electric add on rate')
plt.ylabel('Premium')

plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.hist(df2['No Claim bonus discount'], bins= 20,color='red', edgecolor='black')
plt.xlabel('No Claim Bonus Discount')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.title('Histogram of Premium')
plt.show()
```
