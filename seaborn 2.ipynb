```python

# Import seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset('car_crashes')
tips
```

```python

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="size",hue='day',
  style="sex", size="time"
)
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="day",
    hue="smoker", style="smoker", size="sex",
)
```

```python
dots = sns.load_dataset("dots")
dots
```

```python
sns.relplot(
    data=dots, kind="scatter",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)

#Kernel Density Estimate (KDE) plots provide a smooth representation of the
#probability density function of a single numerical variable
#,highlighting modes and the overall shape of the distribution.
```

```python
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker", rug=True)
```
