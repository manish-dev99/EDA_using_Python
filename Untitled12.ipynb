```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Flipkart.csv')
df.head(50)
```

```python
df['Brand'].value_counts()
```

```python
df.info()
```

```python
brand_counts = df['Brand'].value_counts()
print("Number of unique values for each brand:")
print(brand_counts)
```

```python
sns.set_style('darkgrid')
sns.set_palette('viridis')
sns.set_context('paper')
filtered_df = df[df['Brand'].isin(['POCO'])]
sns.lineplot(x='Model',y='Rating',data=filtered_df,hue='Brand',marker='o')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.violinplot(x='Color',y='Rating',data=filtered_df,hue='Model',split=True)
```

```python
filtered_df2 = df[df['Brand'].isin(['Asus'])]
```

```python
sns.violinplot(x='Color',y='Rating',data=filtered_df2,hue='Model',split=True)
```
