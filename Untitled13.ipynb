```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mobile_sales.csv')
df.head()
```

```python
df.info()
```

```python
sns.set_style('whitegrid')
sns.set_palette('Reds')
sns.set_context('notebook')
sns.barplot(x='Year',y='Units Sold',data=df,color='red')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_palette('Reds')
sns.set_context('notebook')
sns.lineplot(x='Year',y='Units Sold',data=df,color='red',hue='5G Capability',marker='o')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_palette('Reds')
sns.set_context('notebook')
sns.lineplot(x='Quarter', y='Units Sold', data=df, color='green', marker='o')
plt.title('Units Sold by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Units Sold')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Quarter', y='Units Sold', hue='Product Model', data=df)
plt.title('Units Sold by Quarter and Product Model')
plt.xlabel('Quarter')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.legend(title='Product Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df['Product Model'].value_counts()
```

```python
product_model_counts = df['Product Model'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(product_model_counts, labels=product_model_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Product Models')
plt.axis('equal')
plt.show()
```
