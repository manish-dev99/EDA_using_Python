```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mobile_sales.csv')
df.head()
```

```python
units_by_year = df.groupby('Year')['Units Sold'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(units_by_year['Year'], units_by_year['Units Sold'], marker='o')
plt.xlabel('Year')
plt.ylabel('Units Sold')
plt.title('Total Units Sold per Year')
plt.grid(True)
plt.show()
```

```python
samsung_df = df[df['Product Model'].str.contains('Galaxy', case=False)]
samsung_units_by_year = samsung_df.groupby('Year')['Units Sold'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(samsung_units_by_year['Year'], samsung_units_by_year['Units Sold'], marker='o', color='orange')
plt.xlabel('Year')
plt.ylabel('Units Sold (Samsung)')
plt.title('Samsung Units Sold per Year')
plt.grid(True)
plt.show()
```

```python
galaxy_models_units = samsung_df.groupby('Product Model')['Units Sold'].sum().reset_index()
print("Total Units Sold for each Galaxy Model:")
print(galaxy_models_units)
```

```python
plt.figure(figsize=(12, 7))
plt.bar(galaxy_models_units['Product Model'], galaxy_models_units['Units Sold'], color='skyblue')
plt.xlabel('Product Model')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold for each Galaxy Model')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(galaxy_models_units['Product Model'], galaxy_models_units['Units Sold'], color='green')
plt.xlabel('Product Model')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold for each Galaxy Model (Scatter Plot)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
