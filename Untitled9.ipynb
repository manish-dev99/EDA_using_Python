```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
emp_data= {
    "emp_id" : [10121, 10012, 13400, 10011],
    "name" : ['Ansh', 'Aman','Krishna','Nitin'],
    "department" : ['e-commerce', 'computer','agriculture','maintainence'],
    "salary" : [20000,500000,750000,1200020],
    'join_date': ['12/23', '23/25', '14/24', '07/23']
}
df = pd.DataFrame(emp_data)
df
```

```python
filter_salary = df[df['salary'] > 50000]
display(filter_salary)
```

```python
df['bonus'] = df['salary'] * 10 / 100
display(df)
```

```python
average_salary_by_department = df.groupby('department')['salary'].mean()
display(average_salary_by_department)
```

```python
df['join_date'] = pd.to_datetime(df['join_date'], format='%m/%y', errors='coerce')
sorted_df = df.sort_values(by='join_date')
display(sorted_df)
```
