```python
# Count the number of flights for each airline and sort in descending order
airline_counts = df['airline'].value_counts().reset_index()
airline_counts.columns = ['airline', 'flight_count']

# Create a bar plot
ax = sns.barplot(x='airline', y='flight_count', data=airline_counts)

# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)

plt.title('Number of Flights per Airline')
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
